## Deskripsi Proyek
Proyek ini membangun pipeline machine learning end-to-end menggunakan **TensorFlow Extended (TFX)** untuk memprediksi penyakit jantung berdasarkan dataset Heart Disease UCI. Pipeline mencakup seluruh tahap: data ingestion, validasi, transformasi, hyperparameter tuning, pelatihan model, evaluasi, dan deployment.

## 1. Informasi Dataset

Dataset yang digunakan adalah **Heart Disease Dataset** dari UCI Machine Learning Repository. Dataset ini berisi informasi klinis pasien untuk memprediksi keberadaan penyakit jantung.

| Atribut | Deskripsi |
|---|---|
| `age` | Usia pasien (tahun) |
| `sex` | Jenis kelamin (0 = wanita, 1 = pria) |
| `cp` | Tipe nyeri dada (0-3) |
| `trestbps` | Tekanan darah istirahat (mm Hg) |
| `chol` | Kolesterol serum (mg/dl) |
| `fbs` | Gula darah puasa > 120 mg/dl (0/1) |
| `restecg` | Hasil EKG istirahat (0-2) |
| `thalach` | Denyut jantung maksimum |
| `exang` | Angina akibat olahraga (0/1) |
| `oldpeak` | Depresi ST |
| `slope` | Kemiringan segmen ST (0-2) |
| `ca` | Jumlah pembuluh utama (0-4) |
| `thal` | Thalassemia (0-3) |
| `target` | **Label:** 0 = tidak ada penyakit, 1 = ada penyakit |

- **Jumlah sampel:** 303 baris (dataset di sini menggunakan subset 79 sampel untuk demonstrasi)
- **Jumlah fitur:** 13 fitur + 1 target

## 2. Persoalan yang Ingin Diselesaikan

Penyakit jantung merupakan penyebab kematian nomor satu di dunia. Deteksi dini sangat krusial untuk meningkatkan angka keselamatan pasien. Proyek ini bertujuan:

1. Membangun model prediksi penyakit jantung yang akurat dan dapat diandalkan.

## 3. Solusi Machine Learning

Sistem dibangun menggunakan **TensorFlow Extended (TFX)** dengan komponen:

| Komponen | Fungsi |
|---|---|
| `ExampleGen` | Ingesti data CSV |
| `StatisticsGen` | Menghitung statistik data |
| `SchemaGen` | Inferensi skema data |
| `ExampleValidator` | Validasi data terhadap skema |
| `Transform` | Preprocessing & normalisasi fitur |
| `Tuner` | Hyperparameter tuning otomatis (Keras Tuner) |
| `Trainer` | Pelatihan model Neural Network |
| `Resolver` | Memilih model terbaik sebelumnya |
| `Evaluator` | Evaluasi performa model dengan TFMA |
| `Pusher` | Menyimpan model yang lulus evaluasi |

**Target yang ingin dicapai:**
- Binary Accuracy minimal 80%
- Model dapat diakses melalui REST API di cloud
- Sistem dimonitor dengan Prometheus dan Grafana

## 4. Metode Pengolahan Data dan Arsitektur Model

### Pengolahan Data (`modules/transform.py`)
- **Fitur numerik** (`age`, `trestbps`, `chol`, `thalach`, `oldpeak`): diskalakan ke rentang [0, 1] menggunakan `tft.scale_to_0_1`
- **Fitur kategorikal** (`sex`, `cp`, `fbs`, `restecg`, `exang`, `slope`, `ca`, `thal`): digunakan as-is tanpa encoding tambahan
- **Label** (`target`): di-cast ke `int64`

### Hyperparameter Tuning (`modules/tuner.py`)
- Menggunakan **Keras Tuner RandomSearch**
- Parameter yang dituning:
  - `num_layers`: 1–3 hidden layers
  - `units_i`: 32–256 unit per layer (step=32)
  - `learning_rate`: pilihan dari [0.01, 0.001, 0.0001]
- 5 trials untuk menemukan konfigurasi terbaik

### Arsitektur Model (`modules/trainer.py`)
- **Input layer:** 13 fitur (numeric + categorical)
- **Hidden layers:** 1–3 Dense layer dengan aktivasi ReLU (dituning)
- **Output layer:** 1 neuron dengan aktivasi Sigmoid (klasifikasi biner)
- **Optimizer:** Adam
- **Loss:** Binary Crossentropy

### Metrik Evaluasi
- **Binary Accuracy** (threshold: >= 0.5)
- **Example Count**
- Slicing berdasarkan fitur `sex`
